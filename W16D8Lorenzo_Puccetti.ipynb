```python
#Utilizzo il metodo groupby per raggruppare dei dati per i valori di una colonna
# ed il metodo sum per ottenere il numero di casi totali in ogni continente
#Totale_casi_continenti = Covid['new_cases'].sum()

#seguono i dati relativi al numero di casi totali per continente senza aver considerato 
#i dati privi di una localizzazione continentale.
total_cases = Covid.groupby('continent')['new_cases'].sum()
somma_per_continenti = total_cases.sum()
print (somma_per_continenti)
total_cases
```

```python
filtrocont1 = Covid.loc [:, "continent"] == "Africa"
africa = Covid.loc[filtrocont1, "total_cases"]
massimo_africa = africa.max()
media_africa = africa.mean()

filtrocont2 = Covid.loc [:, "continent"] == "Oceania"
oceania = Covid.loc[filtrocont2, "total_cases"]
massimo_oceania= oceania.max()
media_oceania = oceania.mean()

print ("Il massimo di casi totali in Africa è stato "+ str("%.0f"% massimo_africa)+ ", " +
       "il massimo di casi totali in Oceania è stato "+ str("%.0f"% massimo_oceania))


print( "La media di casi totali in Africa è "+ str("%.2f"% media_africa)+ ", " +
       "la media di casi totali in Oceania è "+ str("%.2f"% media_oceania))
```

```python
percentualecont1 = Covid.loc [:, "continent"] == "Africa"
percentuale_africa = Covid.loc[percentualecont1, "new_cases"]
totale_africa = percentuale_africa.sum()


percentualecont2 = Covid.loc [:, "continent"] == "Oceania"
percentuale_oceania = Covid.loc[percentualecont2, "new_cases"]
totale_oceania= percentuale_oceania.sum()
print ("%.2f"%(totale_africa/somma_per_continenti*100))
print("%.2f"%(totale_oceania/somma_per_continenti*100))

#Emerge che i numeri percentuali di casi covid manifestati in Africa ed Oceania sono pari rispettivamente 
#all'1.69% e all'1,93%
#é plausibile che in Africa non ci siano stati i mezzi adeguati per effettuare tamponi ed aggiornare 
#i dati pandemici come invece avvenuto in Europa o in America del Nord.

```

```python
#Dovessimo considerare i casi relativi anche ad aree non catalogate ed ascrivibili ai continenti, la percentuale dei casi totali di questi continenti rispetto 
#ai casi totali mondiali porterebbe ad un risultato fuori scala e fuori dalla realtà. 

#utilizzo la funzione fillna. per sostituire i valori Nan presenti nella colonna dei continenti con Dati_senza_catalogazione_geografica,
#in modo tale da ricalcolare il numero totale di casi Covid come richiesto dalla consegna
Covid['continent'] = Covid['continent'].fillna("Dati_senza_catalogazione_geografica", inplace=False)
total_cases = Covid.groupby('continent')['new_cases'].sum()
somma_per_continenti = total_cases.sum()
print ("La percentuale di casi Covid totali rispetto alla somma dei valori anche dei dati privi di una categorizzazione continentale è pari a " + str(("%.2f"%(totale_africa/somma_per_continenti*100))+ " per l'Africa."))
print("La percentuale di casi Covid totali rispetto alla somma dei valori anche dei dati privi di una categorizzazione continentale è pari a " + str(("%.2f"%(totale_oceania/somma_per_continenti*100))+ " per l'Oceania."))
```

```python
#Riguardo le nazioni di Italia, Germania e 
#Francia, segue il boxplot circa il numero di pazienti in terapia 
#intensiva da maggio 2022 (incluso) ad aprile 2023 (incluso).
#Emerge la netta diminuzione di pazienti ricoverati in terapia intensiva in Italia a dispetto di medie ben più elevate rispetto a Francia e Germania.
#Un'altra differenza si nota dal fatto che sebbene la Germania presenti un numero ben più elevato di ricoveri, 
#la media di pazienti in terapia intensiva è molto simile a quella francese. E' evidente che la Germania abbia subito un volume molto maggiore di pazienti
#ricoverati e quindi una pressione sulle strutture sanitarie maggiore rispetto a Francia ed Italia.

import seaborn as sns
start_date = '2022-05-01'
end_date = '2023-04-30'
GIF2022 = Covid[Covid["location"].isin(["Germany", "Italy", "France"]) & (Covid['date'] >= start_date) & (Covid['date'] <= end_date)]

GIF2022.boxplot(by='location', column=['icu_patients'])
```

```python

#Segue il grafico a barre relativo al numeri di pazienti ospitalizzati nel 2023 relativo alle nazioni di Germania, Italia, Spagna e Francia.
#Nei riguardi della prima si evince che la Germania nel 2023 ha smesso di aggiornare e pubblicare i dati sui pazienti ricoverati a causa del Covid.
#I valori Nan sono quindi stati modificati in "0" per permettere la costruzione del grafico.

import matplotlib.pyplot as plt 

# creo due variabili che utilizzerò per filtrare i dati su una base temporale
data_2023 = "2023-01-01"
data_fine_2023 = "2023-12-31"        

#utilizzo il metodo .fillna() per sostituire i Nan con "0"

GIFS2023 = Covid[Covid["location"].isin(["Germany", "Italy", "France", "Spain"]) & (Covid['date'] >= data_2023) & (Covid['date'] <= data_fine_2023)]
GIFS2023['hosp_patients'].fillna(0, inplace=False)


sum_hosp_patients = GIFS2023.groupby('location')['hosp_patients'].sum()
sum_hosp_patients.plot(kind='bar')
# Assegno etichette e titoli al mio grafico
plt.xlabel("Location")
plt.ylabel("Pazienti ospitalizzati")
plt.title("Pazienti ospitalizzati nel 2023 in Germania, Italia, Francia e Spagna)")
plt.show()
```
